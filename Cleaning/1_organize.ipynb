Add labels - age, sex, height, weight, unit_type, origin into LTM

```python
# Select the needed columns from labels
cols = ["original_uniquepid", "raw_age", "sex", "raw_height", "raw_weight", "unit_type", "origin", "lengthofstay", "intime", "outtime",
       "death_time_from_intime", "mortality"]
labels_subset = labels[cols]

# Merge into ltm based on 'patient'
ltm = ltm.merge(labels_subset, how="left", left_on="patient", right_index=True)
```

```python
ltm
```

```python
ltm.patient.isna().sum()
```

```python
ltm = ltm.rename(columns={"patient": "stay_id", "lengthofstay": "los", "mortality": "icu_mortality", "original_uniquepid": "subject_id"})
```

```python
ltm.to_parquet('../Dataset/mimic-iv/from_pipeline/ltm.parquet')
```
