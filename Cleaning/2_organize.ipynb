- Rename and subsample to 2000.

```python
import pandas as pd
import numpy as np
```

```python
ltm = pd.read_parquet("../Dataset/mimic-iv/from_pipeline/ltm.parquet")
```

```python
ltm
```

```python
ltm.patient.nunique()
```

```python
ltm = ltm.rename(columns={"patient": "stay_id", "lengthofstay": "los", "mortality": "icu_mortality"})
```

```python
ltm
```

```python
# pick 1000 unique stay_ids
stay_sample = (
    ltm["stay_id"]
    .drop_duplicates()
    .sample(n=5000, random_state=42)
)

# filter ltm to only those stay_ids
ltm_sub = ltm[ltm["stay_id"].isin(stay_sample)]
```

```python
ltm_sub.stay_id.isna().sum()
```

```python
ltm_sub.stay_id.nunique()
```

```python
ltm_sub.to_parquet("../Dataset/mimic-iv/from_pipeline/ltm_5000.parquet")
```
